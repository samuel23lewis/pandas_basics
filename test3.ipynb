```python
import pandas as pd

df=pd.read_csv("data1.csv")
print(df)
```

```python
#to fix the wrong data of a perticular col/row ie cell
df.loc[3,'rand']=343
df.loc[95,'rand']=563
print(df)
```

```python
#to replace all the values of 'rand' which is higher than 120
df=pd.read_csv('data1.csv')
for x in df.index:
    if df.loc[x,'rand']>120:
        df.loc[x,'rand']=120
print(df.to_string())
```

```python
#to drop all the values of 'rand' which is higher than 120
df=pd.read_csv('data1.csv')
for x in df.index:
    if df.loc[x,'rand']>120:
        df.drop(x,inplace=True)
print(df.to_string())
```

```python
#to replace any duplicate values
df=pd.read_csv('data1.csv')
df.drop_duplicates(inplace=True)
print(df.to_string())
```
