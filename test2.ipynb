```python
import pandas as pd

df=pd.read_csv('data1.csv')
print(df.to_string())
```

```python
#to display all data with no empty cell
df=df.dropna()
print(df.to_string())
```

```python
#remove all the row with null value
df=pd.read_csv('data1.csv')
df.dropna(inplace=True)
print(df.to_string())
```

```python
#replace all the null value with and number
df=pd.read_csv('data1.csv')
df.fillna(173,inplace=True)
print(df.to_string())
```

```python
#replace null value of perticular [address] with any other value
df=pd.read_csv('data1.csv')
df["address"].fillna(354,inplace=True)
df["rand"].fillna(294,inplace=True)
print(df.to_string())
```

```python
#replace all the null values with mean value of the dataset
x=df["rand"].mean()
df["rand"].fillna(x,inplace=True)
print(df.to_string())
```

```python
#replace all the null values with mean value of the dataset
x=df["rand"].median()
df["rand"].fillna(x,inplace=True)
print(df.to_string())
```

```python
#replace all the null values with mean value of the dataset
x=df["rand"].mode()
df["rand"].fillna(x,inplace=True)
print(df.to_string())
```
